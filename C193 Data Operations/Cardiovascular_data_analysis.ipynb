```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cardiovascular.csv')
df

new_df = df.dtypes
new_df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
group_by_age = df.groupby('age')['cholesterol'].mean().reset_index()
group_by_age
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
label = group_by_age['age']
value = group_by_age['cholesterol']

plt.subplots(figsize=(19,8))

plt.plot(label, value, label="Average Cholesterol in various age groups",linewidth=3.5)

plt.xlabel("Age")
plt.ylabel("Average Cholesterol")
plt.title("AVERAGE CHOLESTEROL IN VARIOUS AGE GROUPS",fontsize=25)
plt.legend()

plt.show()
```

Conclusion - the cholesterol increases as age increases

# Correlation between systolic and diastolic blood pressure

```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats

#predefine code for image
from IPython.display import Image
Image(filename='cardiovascular_img.jpg') 
#predefine code end
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
group_by_systolic = df.groupby('age')['systolic_blood_pressure'].max().reset_index()
group_by_systolic
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it
group_by_diastolic = df.groupby('age')['diastolic_blood_pressure'].max().reset_index()
group_by_diastolic
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
systolic_label = group_by_systolic['age']
systolic_value = group_by_systolic['systolic_blood_pressure']

diastolic_label = group_by_diastolic['age']
diastolic_value = group_by_diastolic['diastolic_blood_pressure']

plt.subplots(figsize=(19,8))

plt.plot(systolic_label,systolic_value,label="Systolic Blood Pressure",linewidth=3.5)
plt.plot(diastolic_label,diastolic_value,label="Diastolic Blood Pressure",linewidth=3.5)

plt.xlabel("Age")
plt.ylabel("Maximum Diastolic & Systolic Blood Pressure")
plt.title("Maximum Diastolic & Systolic Blood Pressure For Various Ages",fontsize=25)
plt.legend()

plt.show()
```

Conclusion - When the diastolic blood pressure increases, the systolic blood pressure decreases. When the diastolic blood pressure decreases, the systolic blood pressure increases. Therefore, we can conclude that systolic and diastolic blood pressures are correlated!
